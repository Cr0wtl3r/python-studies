# Strings em Python


## Indices em strings:

As strings funcionam de certa forma como uma "Lista de caracteres" obedecendo então a indexação de listas propriamente dito. Portanto, o Python interpreta a string como uma lista de caracteres isolados. Exemplo o nome "Albino":

```
A l b i n o
0 1 2 3 4 5

```
Para o Python o nome "Albino" é uma string que nada mais é do que uma lista de 6 caractéres distintos que possuem índices próprios na string. Sempre esses índices começando do 0 da esqueda para direita. Como mostrado acima.

####OBS: OS ESPAÇOS VAZIOS SÃO CONTADOS NOS CARACTERES DA LISTA STRING.


Para pegar um item de uma string usa-se colchetes `[]` com o índice do valor que tu quer pegar (lembrando que começa sempre em 0).

Exemplo:

````python
nome = "Albino"
print(nome[0])
print(nome[1])
print(nome[2])
print(nome[3])
print(nome[4])
print(nome[5])
````

Para descobrir o tamanho de uma String, utiliza-se a função `len()`. Ela retornara quantos caracteres tem a String. Lembrando que espaços são contados.  

````python
nome = 'Albino'
sobrenome = ' dos Santos'
nome_completo = nome + sobrenome
print(len(nome)) # 6 Caracteres!
print(len(nome_completo)) # 17 caracteres, sendo 15 letras e 2 espaços!
````

## Índices negativos e Pedaços de String

Da mesma forma que conseguimos contar de traz pra frente, é possível selecionar os índices da string "de traz pra frente" com índices negativos. Desta forma, o -1 seria o último caractere.
Isso é especialmente útil quando não se sabe qual o índice do último caractere por exemplo, daí para recuperar seu valor, usa-se o -1.
Exemplo:


````python
nome = 'Albino dos Santos'
print(len(nome))
print(nome[-1]) # Será o 's' minúsculo do final do nome.
print(nome[-2])
print(nome[-3])
print(nome[-4])
print(nome[-5])
print(nome[-6])
print(nome[-7])
print(nome[-8])
print(nome[-9])
print(nome[-10])
print(nome[-11])
print(nome[-12])
print(nome[-13])
print(nome[-14])
print(nome[-15])
print(nome[-16])
print(nome[-17])
````

Agora para pegar um pedaço e uma string, podemos fazer da seguinte forma:
```
nome = 'Albino'
print(nome[:5])
```
Onde o retorno desse código será "Albin" pois ele pegara 5 caracteres da String começando do 0.

````python
nome: str = 'Albino'
print(nome[:5])
print(nome[:])
print(nome[2:])
print(nome[2:5])
print(nome[:-1])
````

## Operações com Strings:

* Concatenação: A concatenação é a operação de juntar duas strings em uma única string. O operador de concatenação é o sinal de soma `+`.
* Repetição: A repetição é a operação de repetir uma string um determinado número de vezes. O operador de repetição é o sinal de asterisco `*`.
* Fatiamento: O fatiamento é a operação de extrair uma sub-string de uma string. O operador de fatiamento é o sinal de colchetes `[]`.
* Formatação: A formatação é a operação de inserir valores em uma string. O método `format()` é usado para formatar strings.
* Interpolação de strings (Python 3.6+): A interpolação de strings é uma forma mais recente de formatar strings. Ela usa a letra `f` antes da string e chaves `{}` para indicar onde os valores devem ser inseridos.


#### Métodos

* `capitalize()`: Converte o primeiro caractere da string para maiúscula.
* `lower()`: Converte todos os caracteres da string para minúscula.
* `upper()`: Converte todos os caracteres da string para maiúscula.
* `strip()`: Remove espaços em branco no início e no final da string.
* `find()`: Retorna o índice da primeira ocorrência da substring especificada.
* `replace()`: Substitui todas as ocorrências da substring especificada por outra substring.



### Concatenação:

````python
a = "hello "
b = "world"
c = a + b
print(c)  # Output: hello world
````

### Repetição:

````python
a = "hello"
b = a * 3
print(b)  # Output: hellohellohello
````

### Fatiamento:

````python
a = "hello world"
print(a[0:5])  # Output: hello
print(a[6:])  # Output: world
````

### Formatação:

````python
a = "hello {} {}".format("world", "!")
print(a)  # Output: hello world !
````

### Interpolação de string (Python 3.6+):

````python
a = f"hello {'world'} !"
print(a)  # Output: hello world !
````

### Métodos:

````python
a_string = "hello world  "
print(a_string.capitalize())  # Output: Hello world
print(a_string.lower())  # Output: hello world
print(a_string.upper())  # Output: HELLO WORLD
print(a_string.strip())  # Output: hello world
print(a_string.find("world"))  # Output: 6
print(a_string.replace("world", "universe"))  # Output: hello universe
````

## Operadores de associação

Os operadores de associação são utilizados para verificar se uma sequência contém um objeto. Na Tabela 6 temos os exemplos.

* `in` - Retorna `True` caso o valor seja encontrado na sequência: `2 in x`
* `not in` - Retorna `True` caso o valor não seja encontrado na seuqência: `2 not in x`

Vamos ver um exemplo usando os operadores de associação:




````python
frutas = ["banana","laranja","uva","ameixa"]

fruta_1 = "ameixa"
fruta_2 = "melancia"

print(fruta_1 in frutas) # True
print(fruta_2 in frutas) # False
print(fruta_1 not in frutas) # False
print(fruta_2 not in frutas) # True
````

# Métodos de String em Python:

Primeiramente precisamos diferenciar uma função de um método.

Em Python, tanto funções quanto métodos são blocos de código reutilizáveis que executam tarefas específicas. No entanto, há uma distinção fundamental entre eles:

## Funções:

* **Definição**: Uma função é um bloco de código independente que realiza uma operação específica e retorna um valor (ou None).
* **Sintaxe**: São definidas usando a palavra-chave def, seguida pelo nome da função, parênteses para argumentos (opcionais) e dois pontos.
* **Chamada**: São chamadas diretamente pelo seu nome, seguido por parênteses contendo os argumentos (se houver) `funcao(argumentos)`.
Exemplo:

````python
def saudacao(nome):
    return f"Olá, {nome}!"

mensagem = saudacao("Albino")
print(mensagem)  # Saída: Olá, Alice!
# Observe que o próprio "print()" é uma função assim como "saudacao()".
````

## Métodos:

* **Definição**: Um método é uma função que está associada a um objeto (instância de uma classe). Ele opera sobre os dados do objeto e pode modificar seu estado.
* **Sintaxe**: São definidos dentro da definição de uma classe, usando a mesma sintaxe de uma função, mas com um primeiro parâmetro especial chamado self, que se refere à instância do objeto.
* **Chamada**: São chamados usando a sintaxe `objeto.metodo(argumentos)`.
Exemplo:

````python
class Cachorro:
    def __init__(self, nome, raca):
        self.nome = nome
        self.raca = raca

    def latir(self):
        print(f"{self.nome} da raça {self.raca} está latindo!")

meu_cachorro = Cachorro("Rex","Labrador")
meu_cachorro.latir()  # Saída: Rex da raça Labrador está latindo!
````

- `capitalize()` -> Coloca a 1ª letra Maiúscula

````python

texto = 'albino'
print(texto.capitalize())
# Saída: Albino
````

- `casefold()` -> Transforma todas as letras em minúsculas (existe lower() mas o casefold é melhor normalmente)

````python
nome = 'ALBInO'
print(nome.casefold())
# Saída: albino
````

- `count()`	-> Quantidade de vezes que um valor aparece na string

````python
nome = 'Albino Marques dos Santos'
print(nome.count('a')) # Saída: 2 por que ele é case sensitive e está buscando somente o 'a' minúsculo.
nome_maiúsc = nome.upper()
print(nome_maiúsc.count('A')) # Saída: 3 por que ele é case sensitive e está buscando somente o 'A' maiúsculo.
````

- `endswith()` -> Verifica se o texto termina com um valor específico e dá como resposta True ou False

````python
email = 'albino@gmail.com'
print(email.endswith('@gmail.com')) # Saída: True
print(email.endswith('@gmail')) # Saída: False
print(email.endswith('.com')) # Saída: True
````

- `find()` -> Procura um texto dentro de outro texto e dá como resposta a posição do texto encontrado

````python
print(email.find('@')) # Saída: 6
print(email.find('.')) # Saída: 12
````

- `format()` -> Formata uma string de acordo com os valores passados. Já usamos bastante ao longo do programa.

````python
nome_completo = nome.title()
message = 'Olá, {}!'.format(nome_completo)
print(message) # Saída: Olá, Albino!
````

- `isalnum()` -> Verifica se um texto é todo feito com caracteres alfanuméricos (letras e números) -> letras com acento ou ç são considerados letras para essa função.

````python
nome1 = 'Albino1234'
nome2 = 'Albino'
nome3 = '1234'
print(nome1.isalnum()) # Saída: True
print(nome2.isalnum()) # Saída: True
print(nome3.isalnum()) # Saída: False
````

- `isalpha()` -> Verifica se um texto é todo feito de letras.

````python
texto = 'Albus'
texto1 = 'Albino123'
print(texto.isalpha()) # Saída: True
print(texto1.isalpha()) # Saída: False
````

- `isnumeric()`	-> Verifica se um texto é todo feito por números.

````python
texto3 = '1234'
print(texto3.isnumeric()) # Saída: True
````

- `replace()` -> Substitui um texto por um outro texto em uma string.

````python
text4 = '1000.00'
print(text4.replace('.', ',')) # Saída: 1000,00
print('R$ '+ text4.replace('.',',')) # Saída: R$ 1000,00
````

- `split()`	-> Separa uma string de acordo com um delimitador em vários textos diferentes.

````python
texto5 = 'albino@gmail.com'
print(texto5.split('@')) # Saída: ['albino', 'gmail.com']
````

- `splitlines()` -> separa um texto em vários textos de acordo com os "enters" do texto

````python
texto = '''Olá, bom dia
Venho por meio desse e-mail lhe informar o faturamento da loja no dia de hoje.
Faturamento = R$2.500,00'''
print(texto.splitlines())
# Saída: ['Olá, bom dia', 'Venho por meio desse e-mail lhe informar o faturamento da loja no dia de hoje.', 'Faturamento = R$2.500,00']
````

- `startswith()` -> Verifica se a string começa com determinado texto

````python
cod = 'BEB123456'
print(cod.startswith('BEB')) # Saída: True
print(cod.startswith('BE')) # Saída: True
print(cod.startswith('B')) # Saída: True
print(cod.startswith('0')) # Saída: False
````

- `strip()`	-> Retira caracteres indesejados dos textos do inicio e do fim da string. Por padrão, retira espaços "extras" no início e no final

````python
texto_teste = '   Olá, mundo!   '
texto_teste1 = 'Olá, mundo!?'
print(texto_teste.strip()) # Saída: Olá, mundo!
print(texto_teste1.strip('?')) # Saída: Olá, mundo!
````

- `title()` -> Coloca a 1ª letra de cada palavra em maiúscula

````python
frase = 'O rato roeu a roupa do rei de roma'
print(frase.title()) # Saída: O Rato Roeu a Roupa do Rei de Roma
````

- `upper()`	-> Coloca o texto todo em letra maiúscula

````python
cod2 = 'beb12343'
print(cod2.upper()) # Saída: BEB12343
````

# Método `format()` em Python

O método format em Python é usado para formatar strings, inserindo valores e formatando-os de acordo com especificações dadas. Podendo realizar:

* **Substituição de campos**: `{}` são usados como marcadores de posição para os valores a serem inseridos.

````python
print("Olá, {}!".format("Mundo"))
````

* **Formatação de números**: Pode-se formatar números para ter um número específico de casas decimais, por exemplo:

````python
custo = 500
faturamento = 270
lucro = faturamento - custo
print('Faturamento foi {:.2f} e lucro foi {:.2f}!'.format(faturamento, lucro))

print("O número pi é aproximadamente {:.2f}!".format(3.14159))
````

* **Preenchimento e alinhamento**: É possível especificar o alinhamento (à esquerda, à direita, centralizado) e caracteres de preenchimento.

````python
print("{:*>30}".format("texto"))
print("{:>30}".format("texto"))
print("{:*<30}".format("texto"))
print("{:<30}".format("texto"))
print("{:*^30}".format("texto"))
print("{:^30}".format("texto"))
````

* **Formatação de data e hora**: Com o uso do módulo datetime, pode-se formatar objetos de data e hora.

````python
from datetime import datetime
agora = datetime.now()
print("Data e hora atual: {:%d/%m/%y %H:%M}".format(agora))
````

* **Formatação com palavras chave**: Pode-se usar palavras-chave para inserir e formatar strings de maneira mais legível.

````python
print("{nome} tem {idade} anos".format(nome="João", idade=30))
````

Esses são apenas alguns exemplos. O método format é muito versátil e pode ser personalizado de várias maneiras para atender às necessidades específicas de formatação.

Abaixo mais alguns exemplos:


* **Exemplos de Edição de Sinal**: Sempre que colocado usado para editar, irá acrescentar um sinal sempre ao imprimir a string, sendo ele positivo ou negativo, irá imprimir.

````python
custo = 500
faturamento = 270
lucro = faturamento - custo
print('Faturamento foi {:+} e lucro foi {:+}!'.format(faturamento, lucro))
````

* **Exemplo de Separador de Milhar**: Serve para poder acrescentar na separação do milhar uma virgula para facilitar a identificação.



````python
custo = 5000
faturamento = 2700
lucro = faturamento - custo
print('Faturamento foi {:,} e lucro foi {:,}!'.format(faturamento, lucro))
````

* **Exemplo de Formato Percentual**: O formato percentual permite multiplicar o valor da porcentagem, já adicionar o símbolo de porcentagem (`%`) na saída facilmente e já estipular, caso haja, o número de casas decimais.

**OBS**: Ele já múltiplica por 100 o valor.

````python
custo = 500
faturamento = 1300
lucro = faturamento - custo
margem = lucro / faturamento
print(margem)
print('Faturamento foi {:,} e lucro foi {:,}! A margem foi {:.2%}'.format(faturamento, lucro, margem))
````

* **Exemplo de Formato Moeda - Combinação de Formatos**: Como mencionado antes, é possível implementar uma configuração de format para adequar melhor a string como necessário. Como abaixo:

````python
custo = 5000
faturamento = 27000
lucro = faturamento - custo
print('Faturamento foi R${:,.2f} e lucro foi R${:,.2f}!'.format(faturamento, lucro))

#transformando no formato brasileiro
lucro_texto = 'R${:_.2f}'.format(lucro)
print(lucro_texto.replace('.', ',').replace('_', '.'))
````

* **Função `round()`**: A função `round()` não é advinda do método `format()`, mas ela pode ser aplicada e depois utilizada a saída dela no format. Exmeplo:

````python
imposto = 0.15758
preco = 100
valor_imposto = preco * imposto
valor_imposto_formatado = round(valor_imposto, 2)
print(valor_imposto)
print(valor_imposto_formatado)
````

# Como extrair pedaços de String corretamente



## Forma básica:

````python
precos = "Jan: 25, Fev: 27, Mar: 29"
preco_jan = precos[0:7]
print(preco_jan)
preco_fev = precos[9:-9]
print(preco_fev)
preco_mar = precos[-7:]
print(preco_mar)
````

## Fatiar através de Step:

````python
codigo = "1.2.3.4,5,1,2.3.4,7.9"

pedaco_cod = codigo[::]
print(pedaco_cod)
# Por padrão o step pega sempre de 1 em 1, então deixar assim irá printar tudo.

pedaco_cod = codigo[::2]
print(pedaco_cod)
# Agora, se colocar um valor maior no step, ele irá pegar 1, pular o póximo e pegar outro nesse caso.

pedaco_cod = codigo[::3]
print(pedaco_cod)
# E se fizermos de 3 em 3 ele pega 1 pula 2 e pega o próximo.

pedaco_cod = codigo[::-1]
print(pedaco_cod)
# Aqui, acontece a mesma coisa que no primeiro exemplo, só que ao contrario, de trás pra frente.

pedaco_cod = codigo[::-2]
print(pedaco_cod)
# Nesse a mesma coisa, irá contar de 2 em 2 só que de trás pra frente.

pedaco_cod = codigo[0:10:2]
print(pedaco_cod)
````
