#### 1. Tamanho de strings. Faça um programa que leia 2 strings e informe o conteúdo delas seguido do seu comprimento. Informe também se as duas strings possuem o mesmo comprimento e são iguais ou diferentes no conteúdo.
<pre>
Compara duas strings
String 1: Brasil Hexa 2006
String 2: Brasil! Hexa 2006!
Tamanho de "Brasil Hexa 2006": 16 caracteres
Tamanho de "Brasil! Hexa 2006!": 18 caracteres
As duas strings são de tamanhos diferentes.
As duas strings possuem conteúdo diferente.
</pre>

```python
string_1 = 'Brasil Hexa 2006'
string_2 = 'Brasil Hexa 2006'

if len(string_1) == len(string_2):
    print(f'As duas strings são de tamanhos iguais. {len(string_1)} a primeira e {len(string_2)} a segunda.')
    if string_1 == string_2:
        print('As duas strings possuem conteúdo igual.')
    else:
        print('As duas strings possuem conteúdo diferente.')
else:
    print(f'As duas strings são de tamanhos diferentes o que já demonstra que não possuem o mesmo conteúdo. {len(string_1)} caracteres a primeira e {len(string_2)} caracteres a segunda.')
```

#### 2. Valida e corrige número de telefone. Faça um programa que leia um número de telefone, e corrija o número no caso deste conter somente 7 dígitos, acrescentando o '3' na frente. O usuário pode informar o número com ou sem o traço separador.
<pre>
Valida e corrige número de telefone
Telefone: 461-0133
Telefone possui 7 dígitos. Vou acrescentar o digito três na frente.
Telefone corrigido sem formatação: 34610133
Telefone corrigido com formatação: 3461-0133
</pre>

```python
telefone = '461-0133'
telefone_corrigido = telefone.replace('-', '')
if (len(telefone_corrigido) == 7):
    print(f'O telefone possui {len(telefone_corrigido)} dígitos. Vou acrescentar o digito 3 na frente.')
    telefone_corrigido = '3' + telefone_corrigido
    print(f'O telefone com o dígito acrescentado está ok e é {telefone_corrigido[:4]}-{telefone_corrigido[4:]}!')
elif (len(telefone_corrigido) ==8):
    print(f'O telefone {telefone_corrigido} está certo!')
else:
    print("Digite o telefone correto!")
```
