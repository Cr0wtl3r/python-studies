```python
# Script padrão para iniciar o navegador Chrome com o selenium e abrir a página web específica.
import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
# from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager


service = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=service)

# Abrindo a página de teste
caminho_do_arquivo = os.path.join(os.getcwd(), "index.html")
navegador.get(caminho_do_arquivo)

# Lendo o arquivo Excel com os dados dos processos para poder pesquisar dinamicamente.
df_process = pd.read_excel("Processos.xlsx")



#Função para validade cidade da busca

def cidade_desejada(cidade):
    pagina_desejada = navegador.find_element(By.XPATH, "/html/body/div/div/div/a[1]")
    if cidade == "Distrito Federal":
        pagina_desejada = navegador.find_element(By.XPATH, "/html/body/div/div/div/a[1]")
    elif cidade == "Rio de Janeiro":
        pagina_desejada = navegador.find_element(By.XPATH, "/html/body/div/div/div/a[2]")
    elif cidade == "São Paulo":
        pagina_desejada = navegador.find_element(By.XPATH, "/html/body/div/div/div/a[3]")
    else:
        print(f"Cidade {cidade} não encontrada!")
    return pagina_desejada

def navegador_pesquisa(cidade):
    menu = navegador.find_element(By.CLASS_NAME, "dropdown-menu")
    pagina_desejada = cidade_desejada(cidade)
    time.sleep(1)
    ActionChains(navegador).move_to_element(menu).click().perform()
    time.sleep(1)
    pagina_desejada.click()
    navegador_list = navegador.window_handles
    main_page = navegador_list[0]
    destiny_page = navegador_list[1]
    navegador.switch_to.window(destiny_page)
# Iterando sobre os dados do Excel para cada reu e realizando a consulta.



for reu in df_process.itertuples():
    index, nome, advogado, processo,cidade, status = reu
    navegador_list = navegador.window_handles
    time.sleep(1)
    navegador_pesquisa(cidade)
    time.sleep(1)

    alert = Alert(navegador)
    campo_nome = navegador.find_element(By.ID, "nome")
    campo_advogado = navegador.find_element(By.ID, "advogado")
    campo_processo = navegador.find_element(By.ID, "numero")
    botao_enviar = navegador.find_element(By.CLASS_NAME, "registerbtn")


    campo_nome.clear()
    campo_advogado.clear()
    campo_processo.clear()
    time.sleep(1)
    campo_nome.send_keys(nome)
    campo_advogado.send_keys(advogado)
    campo_processo.send_keys(processo)
    botao_enviar.click()
    time.sleep(1)
    alert.accept()
    WebDriverWait(navegador, 20).until(EC.alert_is_present())
    alert_text = alert.text
    print(f'A consulta do réu {nome} teve o seguinte resultado:', alert_text)
    time.sleep(1)
    alert.accept()
    time.sleep(1)
    campo_nome.clear()
    campo_advogado.clear()
    campo_processo.clear()
    navegador.close()
    navegador.switch_to.window(navegador_list[0])

navegador.quit()



```
