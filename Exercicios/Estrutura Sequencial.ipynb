#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
print("*"*50)
print("Calculadora de salário:")
print("*"*50)

valor_hora = float(input("Quanto você ganha por hora? \n"))
horas_mes = float(input("Quantas horas você trabalha por mês? \n"))

print(f"Você trabalha {horas_mes} horas por mês e cada hora sua vale R${valor_hora}.")
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
print("*"*50)
print("Calculadora de salário:")
print("*"*50)


salario_bruto = valor_hora*horas_mes

print(f"Seu salário bruto é de R${salario_bruto} mensais!")
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
print("*"*50)
print("Calcula o salário com desconto do IR.")
print("*"*50)

salario_desc_ir = (salario_bruto * (11/100))

print(f"O salário com desconto do Imposto de Renda é de {salario_desc_ir}")
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
print("*"*50)
print("Calcula o salário com desconto do INSS.")
print("*"*50)

salario_desc_inss = (salario_bruto * (8/100))

print(f"O salário com desconto do Imposto de Renda é de {salario_desc_inss}")
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
print("*"*50)
print("Calcula o salário com desconto do Sindicado.")
print("*"*50)

salario_desc_sind = (salario_bruto * (5/100))

print(f"O salário com desconto do Imposto de Renda é de {salario_desc_sind}")
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
print("*"*50)
print("Calcula o salário líquido. (Com todos os descontos.)")
print("*"*50)

salario_liquido = salario_bruto - (salario_desc_ir+salario_desc_inss+salario_desc_sind)

print(f"O salário com desconto do Imposto de Renda é de {salario_liquido}")
```

#### 13. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
area = float(input("Qual o tamanho da área em metros quadrados (m²) que irá pintar?"))
# 1L/3m
#Tinta em latas de 18L de R$ 80,00 cada
#Quantas latas necessárias
litros_tinta = area/3

latas = litros_tinta/18

valor = latas*80


print(f"As latas necessárias serão {latas} custando {valor}.")
```

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte.

```python
net = float(input("Quantos Megas são sua internet?"))
tamanho = float(input("Quantos Megas é o arquivo?"))

tempo_seg = tamanho/(net/8)
tempo_min = tempo_seg/60

print(f"Você demoraria {tempo_min}min para baxar o arquivo de {tamanho}MB.")
```
