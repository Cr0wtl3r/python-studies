### Exercício 1
Crie um sistema de consulta de preços
Seu sistema deve:
- Pedir para o usuário o nome de um produto
- Caso o produto exista na lista de produtos, o programa deve retornar o preço do produto como resposta
       - Ex: O produto celular custa R$1500
- Caso o produto não exista na lista de produtos, o programa deve printar uma mensagem para o usuário tentar novamente


```python
precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 2000}

produto_consultado = input("Digite o produto que deseja consultar o preço: ").lower().strip()

if produto_consultado in precos:
    print(f'O produto {produto_consultado} custa R${precos[produto_consultado]}')
else:
    print('Produto não encontrado. Tente novamente.')
```

### Exercício 3
Dada a lista de preços de produtos, uma loja resolveu fazer um reajuste nos preços dos produtos.
calcule o novo valor dos produtos com base nas seguintes regras:
- Preços até 1.000 vão ter um reajuste de 10% (ou seja, o novo preço será 110% do preço atual)
- Preços até maiores que 1.000 até 2.000 vão ter reajuste de 15%
- Preços acima de 2.000 vão ter reajuste de 20%

```python
precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 3000}

print('Preços Antigos:')

for produto, preco in precos.items():
    print(f'O produto {produto.capitalize()} custa R${preco:.2f}.')

for produto, preco in precos.items():
    if preco <= 1000:
        precos[produto] = preco * 1.1
    elif preco <= 2000:
        precos[produto] = preco * 1.15
    elif preco > 2001:
        precos[produto] = preco * 1.2
    else:
        print('Valor inválido')


print("= =" *60)
print('Preços Novos:')

for produto, preco in precos.items():
    print(f'O produto {produto.capitalize()} custa R${preco:.2f}.')


```

### Exercício 4
 - Edite o programa antigo para ter os 2 dicionários, o de preços originais e o de novos preços
 - Em seguida calcule o valor total de reajuste em R$ que teve entre a lista de produtos original e a lista final


```python
precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 3000}
precos_novo = {}

print('Preços Antigos:')

for produto, preco in precos.items():
    print(f'O produto {produto.capitalize()} custa R${preco:.2f}.')

for produto, preco in precos.items():
    if preco <= 1000:
        precos_novo[produto] = preco * 1.1
    elif preco <= 2000:
        precos_novo[produto] = preco * 1.15
    elif preco > 2001:
        precos_novo[produto] = preco * 1.2
    else:
        print('Valor inválido')

print("= =" *60)
print('Preços Novos:')

for produto, preco in precos_novo.items():
    print(f'O produto {produto.capitalize()} custa R${preco:.2f}.')


print("= =" *60)
print('Diferença de valor:')

valor = 0
for precos in precos.values():
    valor += precos

valor_novo = 0
for precos_novo in precos_novo.values():
    valor_novo += precos_novo

valor_diferenca = valor_novo - valor
print(f'O valor de diferença é de R${valor_diferenca:.2f}!')
```


### Exercício 5
- Uma empresa está analisando os resultados de vendas do 1º semestre de 2022 e 2023
- Qual foi o % de crescimento de cada mês de 2023 em relação a 2022?
- Depois de calcular isso, calcule o valor total de crescimento de 2023 em relação a 2022

```python
vendas_22 = {"jan": 15000, "fev": 15500, "mar": 14000, "abr": 16600, "mai": 16300, "jun": 17000}
vendas_23 = {"jan": 17000, "fev": 15000, "mar": 17500, "abr": 16900, "mai": 16000, "jun": 18500}

if len(vendas_22) == len(vendas_23):
    for mes, preco in vendas_22.items():
        if mes in vendas_23:
            crescimento = ((vendas_23[mes] - vendas_22[mes]) / vendas_22[mes])
            print(f'O mês {mes.capitalize()} teve um crescimento de {crescimento:.2%} em relação ao ano passado!') if crescimento > 0 else print(f'O mês {mes.capitalize()} teve um queda de {crescimento:.2%} em relação ao ano passado!')
        else:
            print(f'O mês {mes.capitalize()} não foi encontrado no ano passado!')

sum_22 = sum(vendas_22.values())
sum_23 = sum(vendas_23.values())

print("= =" *60)

print(f'O valor total de crescimento foi de {((sum_23 - sum_22) / sum_22):.2%} em relação ao ano passado!')
```
