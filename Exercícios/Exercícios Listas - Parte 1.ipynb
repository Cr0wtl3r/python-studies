## 1. Faturamento do Melhor e do Pior Mês do Ano

Qual foi o valor de vendas do melhor mês do Ano?
E valor do pior mês do ano?

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]
vendas_total = vendas_1sem + vendas_2sem
total_vendas = sum(vendas_total)
melhor_venda = max(vendas_total)
pior_venda = min(vendas_total)
porcentagem_melhor_venda = (melhor_venda*100)/total_vendas

i_melhor_venda = vendas_total.index(melhor_venda)
i_pior_venda = vendas_total.index(pior_venda)

melhor_venda_formatada = f"{melhor_venda:_.2f}"
pior_venda_formatada = f"{pior_venda:_.2f}"
total_vendas_formatada = f"{total_vendas:_.2f}"
porcentagem_melhor_venda_formatada = f"{porcentagem_melhor_venda:_.2f}"

print(f"O valor do melhor mês do ano foi R${melhor_venda_formatada.replace('.',',').replace('_', '.')}!")
```

## 2. Continuação

Agora relacione as duas listas para printar 'O melhor mês do ano foi {} com {} vendas' e o mesmo para o pior mês do ano.

Calcule também o faturamento total do Ano e quanto que o melhor mês representou do faturamento total.

Obs: Para o faturamento total, pode usar a função sum(lista) que soma todos os itens de uma lista

```python
print(f"O valor de vendas do melhor mês do ano foi de R${melhor_venda_formatada.replace('.',',').replace('_', '.')}, respectivo ao mês {meses[i_melhor_venda]}.")
print(f"O valor de vendas do pior mês do ano foi de R${pior_venda_formatada.replace('.',',').replace('_', '.')}, respectivo ao mês {meses[i_pior_venda]}.")

print(f"O faturamento total do ano foi de R${total_vendas_formatada.replace('.',',').replace('_', '.')}")
print(f"O melhor mês representa o equivalente a {porcentagem_melhor_venda_formatada.replace('.',',').replace('_', '.')}% do faturamento!")
```
