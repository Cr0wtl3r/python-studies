## Oferecendo desconto para cliente com base na última compra

Suponha que você está trabalhando para uma empresa que deseja rastrear a atividade do cliente. Uma métrica que eles estão interessados é o tempo que passou desde a última transação do cliente. Se for muito tempo, eles podem oferecer um desconto para o cliente. Crie um script Python que mostra quanto tempo se passou desde a última compra do cliente. Se faz mais de 30 dias, mostre uma mensagem oferecendo um desconto para o cliente.

```python
from datetime import datetime, timedelta

ultima_compra = input("Digite a data da última compra (dd/mm/aaaa): ")
data_atual = datetime.now()

if ultima_compra:
    ultima_compra = datetime.strptime(ultima_compra, "%d/%m/%Y")
    dias_desde_ultima_compra = (data_atual - ultima_compra).days
    if dias_desde_ultima_compra > 30:
        print(f"Você tem {dias_desde_ultima_compra} dias desde a última compra. Você pode oferecer um desconto!")
    else:
        print(f"Você tem {dias_desde_ultima_compra} dias desde a última compra.")
else:
    print("Você não fez nenhuma compra ainda.")
```


## Data e hora em diferentes fusos horários

Uma empresa tem escritórios em São Paulo, Nova York e Tóquio. Crie um script Python que mostra a data e hora atuais nesses três fusos horários. Exiba, também, se estes escritórios estão abertos ou fechados (9h às 17h).


```python
from datetime import datetime, timedelta
from zoneinfo import ZoneInfo

hora_atual = datetime.now()
hora_atual_sp = hora_atual.astimezone(ZoneInfo('America/Sao_Paulo'))
hora_atual_ny = hora_atual.astimezone(ZoneInfo('America/New_York'))
hora_atual_tokyo = hora_atual.astimezone(ZoneInfo('Asia/Tokyo'))

print(hora_atual)
print(hora_atual_sp)
print(hora_atual_ny)
print(hora_atual_tokyo)

if hora_atual_sp.hour >= 9 and hora_atual_sp.hour < 17:
    print("São Paulo está aberto")
else:
    print("São Paulo está fechado")
if hora_atual_ny.hour >= 9 and hora_atual_ny.hour < 17:
    print("Nova York está aberto")
else:
    print("Nova York está fechado")
if hora_atual_tokyo.hour >= 9 and hora_atual_tokyo.hour < 17:
    print("Tóquio está aberto")
else:
    print("Tóquio está fechado")
```
